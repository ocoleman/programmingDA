# Creating time series

```python
#Adapted from: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
#72 hours starting with midnight Jan 1st, 2011
rng = pd.date_range("1/1/2011", periods=72, freq="H")
```

```python
rng
```

```python
import numpy as np
```

```python
#using rng as the index for a dataset
ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=["admissions"])
```

```python
ts
```

```python
np.random.poisson(5, 10)
```

# Accessing and grouping

```python
ts.mean()
```

```python
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00' ]
```

```python
ts.loc['2011-01-02 00:00:00':'2011-01-02 23:00:00'].mean()
```

```python
ts.resample('D').mean() #aggregate by day
```
